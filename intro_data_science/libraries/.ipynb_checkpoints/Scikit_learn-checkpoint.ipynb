### Exemplos de testes

Dividir um conjunto de dados em dados de treinamento e teste é muito fácil com o sklearn. Tudo que precisamos fazer é usar o comando train_test_split. A função recebe como entradas X, y e test_size (tamanho da amostra), retornando quatro coisas:

 - X_train: a entrada de treinamento
 - X_test: a entrada de teste
 - y_train: os labels de treinamento
 - y_test: os labels de teste

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```
