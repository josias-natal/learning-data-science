# **O que é um Jupyter Notebook?**


   O jupyter notebook e JupyterLab são uma aplicações tanto para **desenvolver e executar**  códigos quanto elementos de texto, figuras, links, vídeos, equações, etc. Como podemos misturar texto e código em um mesmo lugar, estes documentos ficam ideais para gerar e descrever análises onde tanto a formulação da ideia, a visualização do resultado e a descrição do resultado pode ser feita num mesmo lugar.

 - [Documentação do Jupyter](https://jupyter.org/documentation)
 - [Documentação do IPython API](https://ipython.readthedocs.io/en/stable/api/index.html)
 - [Site oficial do Jupyter](http://jupyter.org/)

## **Células**

Isso é uma célula de texto. Células de texto aceitam marcação conhecida como Markdown.
Em uma célula de código nós podemos usar o Python como usamos em um script qualquer:

```python
# isso é uma célula de código

for count in range(0,5):
    print(count)
print('end')
```



Cada célula é um ambiente **isolado** que pode ser rodado separadamente. Você pode rodar cada célula individualmente e na sequência que desejar, mas as variáveis que ela usa precisam estar referenciadas antes do seu uso.

**OBS:** caso haja dependencia entre celulas, garanta que todas as celulas foram executadas (*run all cells*)

```python
import this
```

## **Teclas de atalho**

 - **Tab** = code completion or indent
 - **Shift + Tab** = tooltip
 - **Ctrl + Enter** = run cell
 - **Y** = code
 - **M** = markdown

Também é possível usar fórmulas junto com o texto. Isso é muito útil quando se tem uma explicação matemática necessária para entender o código. A escrita é feita da mesma forma que no LaTeX:
